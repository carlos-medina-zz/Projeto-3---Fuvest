```python
# Cópia das minhas respostas para outra variáveis para deixá-la intacta caso necessário
answer = my_answer.copy()
counter = 0

# Nova lista com número das questões, minha resposta, resposta correta e matéria
for item in test_answer:
    answer[counter].append(item[1])
    answer[counter].append(item[2])
    counter += 1

# Label das colunas
columns = ["Questão", "Minha resposta", "Resposta correta", "Matéria"]
```

```python
# Cria o dataframe
df = pd.DataFrame(np.array(answer), columns = columns)

# Cria uma nova coluna com True caso tenha acertado a questão e False caso tenha errado
df["Acertei questão"] = np.where(df["Minha resposta"] == df["Resposta correta"], True, False)
```

```python
subjects = ["Matemática", "História", "Português", "Literatura", "Química",
            "Biologia", "Física", "Inglês", "Geografia"]

# Lista de listas que irá receber em cada elemento:
# Nome da matéria, quantidade de questões, quantidade de acertos, porcentagem de acertos
grades_list = [ [] for i in range(len(subjects)) ]

counter = 0
for subject in subjects:
    
    # Primeiro elemento recebe o nome da matéria
    grades_list[counter].append(subject)
    
    # Segundo elemento recebe a quantidade de questões
    df_subject = df.query('Matéria == @subject')
    len_subject = len(df_subject)
    grades_list[counter].append(len_subject)
    
    # Terceiro elemento recebe a quantidade de acertos
    mask = df_subject["Acertei questão"] == True
    rights_subject = len(df_subject[mask])
    grades_list[counter].append(rights_subject)
    
    # Quarto elemento recebe a porcentagem de acertos
    percentage_subject = int(100 * (rights_subject/len_subject))
    grades_list[counter].append(percentage_subject)
    
    counter += 1
```

```python
total_rights = 0

# Print dos acertos, total de questões e porcentagem de acertos de cada matéria

print("Dados da minha Fuvest 2020\n")

for item in grades_list:
    print("{0}: {1} de {2} = {3}%" .format(item[0], item[2], item[1], item[3]))
    total_rights = total_rights + item[2]

print("\nAcertei {} questões de {}." .format(total_rights, len(df)))
```
